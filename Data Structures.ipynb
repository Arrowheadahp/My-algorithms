```python
class node:
    def __init__(self, data = None, nxt = None, prv = None):
        self.data = data
        self.nxt = nxt
        self.prv = prv

class linkedlist:
    def __init__(self):
        self.head = node()
        self.tail = self.head
    
    def append(self, data):
        self.tail.nxt = node(data, None, self.tail)
        self.tail = self.tail.nxt
    
    def pop(self):
        self.tail = self.tail.prv
        self.tail.nxt = None
    
    def search(self, data):
        n = self.head
        i = 0
        while n.nxt:
            n = n.nxt
            if n.data == data:
                return i
            i += 1
        return -1
    
    def sort(self):
        i = self.head.nxt
        while i.nxt:
            j = i
            k = i.data
            while j.prv and j.prv.data >k:
                j.data = j.prv.data
                j = j.prv
            j.data = k
        return self
    
    def add(self, data, i):
        n = self.head
        for j in range(i):
            n = n.nxt
        a = node(data, n, n.prv)
        n.prv.nxt = a
    
    def delete(self, data):
        n = self.head
        while n.data != data:
            n = n.nxt
        n.prv.nxt = n.nxt
        
    def __str__(self):
        n = self.head
        s = ''
        while n.nxt:
            s += str(n.data)+ '->'
        return s
```
