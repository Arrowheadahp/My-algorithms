```python
class node:
    def __init__(self, data = None, nxt = None, prv = None):
        self.data = data
        self.nxt = nxt
        self.prv = prv

class linkedlist:
    def __init__(self):
        self.head = node()
        self.tail = self.head
    
    def append(self, data):
        self.tail.nxt = node(data, None, self.tail)
        self.tail = self.tail.nxt
    
    def pop(self):
        t = self.tail.data
        self.tail = self.tail.prv
        self.tail.nxt = None
        return t
    
    def search(self, data):
        n = self.head
        i = 0
        while n.nxt:
            n = n.nxt
            if n.data == data:
                return i
            i += 1
        return -1
    
    def sort(self):    #we use insertion sort here
        i = self.head.nxt
        while i.nxt:
            i = i.nxt
            j = i
            k = i.data
            while j.prv and j.prv.data >k:
                j.data = j.prv.data
                j = j.prv
            j.data = k
        return self
    
    def add(self, data, i):
        n = self.head
        for j in range(i+1):
            n = n.nxt
        a = node(data, n, n.prv)
        n.prv.nxt = a
        n.prv = a
    
    def delete(self, pos):
        n = self.head
        i=-1
        while i<pos :
            n = n.nxt
            i+=1
            if not n.nxt:
                return
        n.prv.nxt = n.nxt
        n.nxt.prv = n.prv
        
    def __str__(self):
        n = self.head
        s = ''
        while n.nxt:
            n = n.nxt
            s += str(n.data)+ '-> '
        return s

LL = linkedlist()
i = 1
while i:
    i = int(raw_input('Enter 1 for append, 2 for pop, 3 to add, 4 to delete, 5 to sort, 6 to search: '))
    if i==1:
        LL.append(input("Enter number to append: "))
        
    elif i==2:
        print LL.pop()
    
    elif i==3:
        LL.add(input('Enter number to add: '), input('Enter position: '))
    
    elif i==4:
        LL.delete(input('Enter position to delete: '))
    
    elif i==5:
        LL.sort()
    
    elif i == 6:
        print LL.search(input('Enter number to search: '))
    
    print LL
```
