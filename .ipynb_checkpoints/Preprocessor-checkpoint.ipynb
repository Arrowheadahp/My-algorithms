```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_boston
R, t = load_boston(True)
R = pd.DataFrame(R)
raw_train_data, raw_test_data, y, v = train_test_split(R, t, random_state=47)
raw_train_data.head(5)
```

```python
raw_train_data.info()
```

```python
raw_train_data.describe()
```

```python
def prepare(raw_DF):
    DF = raw_DF.copy()
    if 'y' in DF.columns:
        DF.drop(columns=['y'])
    
    return DF
```

```python
train_X = prepare(raw_train_data)
test_X = prepare(raw_test_data)

F = train_X.copy()
F['y'] = y
F.head()
```

```python
F.info()
```

```python
F.corr().abs().sort_values('y', ascending=False).y
```

```python
dt = DecisionTreeRegressor().fit(train_X, y)
plt.figure(figsize = (14, 9))
plot_tree(dt, 
          max_depth = 3, 
          filled = True, 
          fontsize=12, 
          label='root', 
          impurity=False, 
          feature_names=train_X.columns
         )
plt.title('Features')
plt.show()
```

```python
rfecv = RFECV(DecisionTreeRegressor())
rfecv.fit(raw_train_data, y)
```
