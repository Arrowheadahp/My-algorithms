```python
# Insertion Sort

def insertionsort(L):
    for i in range(1,len(L)):
        j = i
        k = L[i]
        while j>0 and L[j-1]>k:
            L[j] = L[j-1]
            j -= 1
        L[j]=k
    return L

sortcheck(insertionsort)
```

```python
# bubble sort

def bubblesort(L):
    l = len(L)
    for i in range(l-1):
        for j in range(l-i-1):
            if L[j] > L[j+1]:
                L[j], L[j+1] = L[j+1], L[j]
    return L

sortcheck(bubblesort)
                
```

```python
# Merge Sort

def mergesort(L):
    l = len(L)
    if l <= 1:
        return L
    
    A = mergesort(L[:l/2])
    B = mergesort(L[l/2:])
    S = []
    
    a, b = 0, 0
    while a<len(A) and b<len(B):
        if A[a] < B[b]:
            S.append(A[a])
            a += 1
        else:
            S.append(B[b])
            b += 1
    S += A[a:] + B[b:]
    return S

sortcheck(mergesort)
```

```python
# Quick Sort

from random import choice
def quicksort1(L):
    if len(L) <= 1:
        return L
    k = choice(L)
    A, B, C = [], [], []
    for i in L:
        if i < k:
            A.append(i)
        elif i == k:
            B.append(i)
        elif i > k:
            C.append(i)
    return quicksort1(A) + B + quicksort1(C)


sortcheck(quicksort1)
```

```python
# In Place Quick Sort

def quicksort(L, start = 0, end = None):
    if end == None:
        end = len(L)
    if end - start <= 1:
        return
    #print L[start: end]
    a, b = start+1, start+1
    k = L[start]
    while b < end:
        if L[b] <= k:
            L[a], L[b] = L[b], L[a]
            a += 1
        b += 1
    L[start], L[a-1] = L[a-1], L[start]
    a -= 1
    quicksort (L, start, a)
    quicksort (L, a+1, end)
    return L
    
sortcheck(quicksort)
```
